```python
# Load data Tweets
tweets=pd.read_excel('/content/drive/MyDrive/Tweets Testing Model fix.xlsx')
```

```python
tweets
```

```python
tweets.shape
```

```python
import pandas as pd
import string
import ast

def convert_to_list(text):
    return ast.literal_eval(text)

# Mengubah teks list menjadi list objek Python dalam kolom 'words'
tweets['rmv_punc'] = tweets['text_preprocessed'].apply(convert_to_list)
```

```python
tweets
```

### Sentiment Count

```python
import seaborn as sns
sns.countplot(x='Sentiment', data=tweets)
```

### Word Cloud

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords=set(STOPWORDS)

def wordcloud(data, title=True):
  wordcloud=WordCloud(
      background_color='white',
      stopwords=stopwords,
      max_words=250,
      max_font_size=30,
      scale=2,
      random_state=5
  ).generate(str(data))
  fig=plt.figure(1, figsize=(15, 15))
  plt.axis('off')
  if title:
    fig.suptitle(title, fontsize=20)
    fig.subplots_adjust(top=2.3)
  plt.imshow(wordcloud)
  plt.show()

wordcloud(tweets['rmv_punc'], title='WordCloud Tweets')
```

```python
# Visualize word cloud

list_words=''
for tweet in tweets['rmv_punc']:
  for word in tweet:
    list_words+=' '+(word)

wordcloud=WordCloud(width=600, height=400, background_color='black',
                    min_font_size=10).generate(list_words)
fig, ax=plt.subplots(figsize=(8, 6))
ax.set_title('Word Cloud of Tweets Pilpres', fontsize=18)
ax.grid(False)
ax.imshow((wordcloud))
fig.tight_layout(pad=0)
ax.axis('off')
plt.show()
```

```python
positive_tweets = ' '.join(tweets[tweets['Sentiment'] == 'positif']['rmv_punc'].apply(lambda x: ' '.join(x)))

# Membuat WordCloud dari teks dengan sentimen positif
wordcloud = WordCloud(width=600, height=400, background_color='black',
                      min_font_size=10).generate(positive_tweets)

# Menampilkan WordCloud
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Word Cloud of Positive Sentiment Tweets', fontsize=18)
ax.grid(False)
ax.imshow(wordcloud)
fig.tight_layout(pad=0)
ax.axis('off')
plt.show()
```

```python
negative_tweets = ' '.join(tweets[tweets['Sentiment'] == 'negatif']['rmv_punc'].apply(lambda x: ' '.join(x)))

# Membuat WordCloud dari teks dengan sentimen positif
wordcloud = WordCloud(width=600, height=400, background_color='black',
                      min_font_size=10).generate(negative_tweets)

# Menampilkan WordCloud
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Word Cloud of Negative Sentiment Tweets', fontsize=18)
ax.grid(False)
ax.imshow(wordcloud)
fig.tight_layout(pad=0)
ax.axis('off')
```
